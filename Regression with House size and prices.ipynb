```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# read data from file
data = pd.read_csv('home_dataset.csv')
# extract data
sizes = data['HouseSize'].values
prices = data['HousePrice'].values

#plot of data
plt.scatter(sizes, prices, marker = 'o', color = 'blue')
plt.title('Comparison of house sizes and prices')
plt.xlabel('House size')
plt.ylabel('House price')
plt.show()


# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(sizes, prices, test_size=0.2, random_state=42)

# Reshape the data for NumPy
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)

# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)

# prediction part
predictions = model.predict(x_test)

# Visualize the predictions
plt.scatter(x_test, y_test, marker='o', color='blue', label='Actual Prices')
plt.plot(x_test, predictions, color='red', linewidth=2, label='Predicted Prices')
plt.title('Property Price Prediction with Linear Regression')
plt.xlabel('House Size ')
plt.ylabel('House Price (millions)')
plt.legend()
plt.show()
```
